```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("mesh_Arrays.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
plt.show()
```

```python
shape = np.shape(img)
collection = []
commit = False

for i in range(shape[0]):
    for j in range(shape[1]): 
        commit = (img[i,j] > 100)
        if commit == True:
            collection.append([i,j])
```

```python
collection
```

```python
distance = ((collection[0][0] - collection[1][0])**2  + (collection[0][1] - collection[1][1])**2)**0.5
```

```python
counter = 1
new_collection = []

for i in collection:
    distance = ((i[0] - collection[counter][0])**2  + (i[1] - collection[counter][1])**2)**0.5
    counter += 1
    if distance > 1:
        new_collection.append(i)
    print(distance)
    if counter == len(collection):
        break
```

```python
counter = 1
filtered_collection = []
commit = True


for i in new_collection:
    print(f"collenction with counter { new_collection[counter]} | i values : {i[0]}")
    if commit == True:
        filtered_collection.append(i)
    
    
    if i[0] == new_collection[counter][0]:
        commit = False
    else: 
        print("eit deil arkadam.")
        commit = True
    counter += 1    
    
    if counter == len(new_collection):
        break
```
