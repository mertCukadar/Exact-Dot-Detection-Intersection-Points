```python
import cv2 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
original_img = cv2.imread("hyttest2.png") # call original image

gray_scaled_original_image = cv2.cvtColor(original_img , cv2.COLOR_BGR2GRAY) # converting the original image to gray scale


#ploting proccess
plt.subplot(1,2,1)
plt.imshow(original_img)
plt.title("Original Image")


plt.subplot(1,2,2)
plt.imshow(gray_scaled_original_image)
plt.title("gray_scaled Image")



plt.show()
```
