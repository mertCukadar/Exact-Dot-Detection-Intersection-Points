````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import cv2
````

````python
img = cv2.imread("mesh_Arrays.jpeg")
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
````

````python
shape = np.shape(gray)
print(f"shape of image:\n {shape}")
````

````python
l1 = [] # Intensity > 140
l2 = []
l3 = list(gray)
l4 =[]
````

````python
gray[0,0]
````

````python

for i in range(shape[0]):
    for j in range(shape[1]):
        control = (gray[i,j] > 140)
        if control == True:
            l1.append([i,j])
````

````python
l1[0]
````

````python
l2 = l1.copy()
index = 1

for i in l1:
      
    if  index == len(l1):
        break
    
    
    distance = np.sqrt((i[0] - l1[index][0])**2 + (i[1] - l1[index][1])**2)
    index += 1 
    print(distance)
    if distance > 30:
        l4.append(i)
    else:
        continue
  
        
````

````python
distance
````

````python
l1[3][1]
````

````python
l4
````

````python
l5 = np.array(l4)
````

````python
shape = np.shape(l4)
````

````python
l6 = []
l6.append([l4[0]])
for i in range(len(l4)):
    distance = l4[i][0] -l6[-1][0][0] 
    if distance == 0: 
        l6.append([l4[i]])
    if distance > 10:
        l6.append([l4[i]])
````

````python
l6
````

````python
blank_img = np.zeros(np.shape(gray))
````

````python
np.shape(blank_img)
````

````python
for i in l6:
    print(i[0])
    blank_img[i[0][0],i[0][1]] = 200
    print("it's done")
````

````python
plt.imshow(blank_img)
plt.show()
plt.imsave("blank_img.jpeg",blank_img)
````

````python
original_img = cv2.imread("hyttest2.png")
original_img = cv2.cvtColor(original_img , cv2.COLOR_BGR2GRAY)
original_img = cv2.blur(original_img , (10,10))
plt.imshow(original_img)
plt.show()
````

````python
blank_dot = np.zeros(np.shape(gray))
````

````python
for i in l6:
    rangex_min = i[0][0] - 20
    rangex_max = i[0][0] + 20
    
    rangey_min = i[0][1] - 20
    rangey_max = i[0][1] + 20
        
    region = original_img[rangex_min : rangex_max , rangey_min : rangey_max]
    
    Highest_I = np.transpose(np.where(region == np.min(region)))
    mid = np.floor(len(Highest_I)/2)
    potision = Highest_I[int(mid)]
    
    
    #implament blank dot
    blank_dot[rangex_min : rangex_max , rangey_min : rangey_max][potision[0] , potision[1]] = 255
    
    
    
    
    
    #plt.imshow(region)
    #plt.show()
````
